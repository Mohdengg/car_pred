```python
df_final_dataset=df[[ 'Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
df_final_dataset.head()
```

```python
df_final_dataset['Current Year'] = 2021
```

```python
df_final_dataset['Years used'] = df_final_dataset['Current Year']- df_final_dataset['Year']
```

```python
df_final_dataset.head()
```

```python
df_final_dataset.drop(['Year','Current Year'],axis=1,inplace=True)
```

```python
df_final_dataset=pd.get_dummies(df_final_dataset,drop_first=True)
```

```python
X = df_final_dataset.iloc[:,1:]#independent
y = df_final_dataset.iloc[:,0] #dependent features
```

```python
X.head()
```

```python
y.head()
```

```python
#feature importance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor().fit(X,y)
```

```python
model
```

```python
print(model.feature_importances_)
```

```python
feat_import = pd.Series(data=model.feature_importances_,index = X.columns)
feat_import.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_random = RandomForestRegressor()
```

```python
#Hyper Parameters
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]
print(n_estimators)
```

```python
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, 
                               param_distributions = random_grid,
                               scoring='neg_mean_squared_error',
                               n_iter = 10,
                               cv = 5, 
                               verbose=2, 
                               random_state=42,
                               n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
predictions= rf_random.predict(X_test)
```

```python
predictions
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```

```python
import pickle
file = open('random_forest_regression_model.pkl','wb')

pickle.dump(rf_random,file)
```
